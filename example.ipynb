# Example Jupyter Host File Picker

This notebook demonstrates the basic usage of the Jupyter Host File Picker widget.

```python
# For development, uncomment these lines to hot-reload the javascript code:
# %load_ext autoreload
# %autoreload 2
# %env ANYWIDGET_HMR=1
```

```python
from jupyter_host_file_picker import HostFilePicker
```

```python
picker = HostFilePicker()

selected = []
def record_selected(change):
    """Store selected paths in the `selected` list."""
    selected.clear()
    selected.extend(change['new'])
picker.observe(record_selected, names='selected')

picker
```

After the user selected a file, read the path like this:

```python
picker.selected
```

Or use the list filled by the observer:

```python
selected
```
